### Dimensionality Reduction - Jaffe

```python
import os
import pandas as pd
from PIL import Image
```

```python
directory_path = '../datasets/jaffe/'
expansion = "jaffe"
emotion_folders = ['']
```

```python
def analyze_faces_in_directory(directory, emotion_folder, expansion="", demo=False):
    # Loop through all files in the directory
    size = (48, 48)
    files = os.listdir(directory)
    for demo_i, filename in enumerate(files):
        if demo and demo_i >= 5:
            break
        if filename.endswith(".tiff"):  # Add more extensions if needed
            file_path = os.path.join(directory, filename)
            folder_path = os.path.dirname(file_path)  # Extract folder path
            output_dir = os.path.join(folder_path, "resized_70")
            os.makedirs(output_dir, exist_ok=True)  # Ensure the output directory exists
            output_path = os.path.join(output_dir, filename)
            print(file_path)
            try:
                with Image.open(file_path) as img:
                    # Resize the image
                    # there are several method for resizing, include it in the descriptiomn 
                    METHOD = Image.Resampling.LANCZOS
                    # METHOD = Image.Resampling.BICUBIC
                    # METHOD = Image.Resampling.BILINEAR
                    # METHOD = Image.Resampling.BOX
                    # METHOD = Image.Resampling.HAMMING
                    img_resized = img.resize(size, METHOD)  # Updated Resampling method
                    # Save the resized image to the output directory
                    img_resized.save(output_path, format="TIFF")
                    print(f"Resized and saved: {output_path}")
            except Exception as e:
                # Handle the error gracefully and continue processing
                print(f"Error processing {filename}: {e}")

DEMO = False
emotion_folders = ['']
directory_path = "../datasets/jaffe"  # Replace with the actual dataset path
for e in emotion_folders:
    analyze_faces_in_directory(directory=f"{directory_path}/{e}", 
                               emotion_folder=e,
                               expansion="",
                               demo=DEMO)
```
