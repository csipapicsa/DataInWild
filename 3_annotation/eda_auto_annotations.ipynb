### Exploratory data analysis for automated annotations

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
automated_file = "AutomatedAnnotations_ALL.csv"
manual_file = ""
file = automated_file

df = pd.read_csv(file)
df.head()
```

### Histograms for all the data

```python
interest_columns = ["Age", "Gender", "Race"]

fig, axes = plt.subplots(1, 3, figsize=(14, 5))
for axis, data in zip(axes, [df.Age, df.Gender, df.Race]):
    sns.histplot(data, ax=axis)
```

### Histograms grouped by dataset

```python
dataset_grouped = df.groupby('datasetName')
fig, axes = plt.subplots(len(dataset_grouped), 3, figsize=(16, 12))
fig.tight_layout(pad=5)

for i, (name, group) in enumerate(dataset_grouped):
    for j, (axis, col) in enumerate(zip(axes[i], interest_columns)):
        sns.histplot(group[col], ax=axis)
        if j == 1:
            axis.title.set_text(f"Dataset: {name}")

print(f"Using dataset: {file}")
```
